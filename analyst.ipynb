## fill the null value with mean median or mode // mode for categorical data median and mean for numerical 

```python
def imput_median(dataframe ):
    return dataframe.fillna(dataframe.median)
    
```

```python
google_data.Rating=google_data['Rating'].transform(imput_median)
```

# lambda arguments : expression

```python
### Let's convert Price, Reviews and Ratings into Numerical Values
google_data['Price'] = google_data['Price'].apply(lambda x: str(x).replace('$', '') if '$' in str(x) else str(x))

```

```python
google_data['Price']=pd.to_numeric(google_data['Price'],errors='coerce')
```

```python
google_data.dtypes
```

```python
google_data['Reviews']=pd.to_numeric(google_data['Reviews'],errors='coerce')
google_data['Installs']=google_data['Installs'].apply( lambda x: str(x).replace('+','') if '+' in str(x) else str(x))
google_data['Installs']=google_data['Installs'].apply(lambda x: str(x).replace(',','') if ',' in str(x) else str(x))
google_data['Installs']=pd.to_numeric(google_data['Installs'],errors='coerce')
```
